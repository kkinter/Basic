```python
import openpyxl
import requests
import urllib.request
from bs4 import BeautifulSoup as bs
import pandas as pd
import os
import re


uni_list =[

]


count = 0
li = []
if os.path.isdir("big") == False:
    os.mkdir("big")
# if os.path.isdir("medium") == False:
#     os.mkdir("medium")
# if os.path.isdir("small") == False:
#     os.mkdir("small")
# if os.path.isdir("tiny") == False:
#     os.mkdir("tiny")
if os.path.isdir("excel/old_rep") == False:
    os.mkdir("excel")
    os.mkdir("excel/old_rep")

for uni in uni_list: # < 반복문으로 돌려서,
             
    uni_dir = re.sub('[^a-zA-Z]', '', uni)
#     if os.path.isdir(uni_dir) == False:
#         os.mkdir(uni_dir)  # 폴더 생성

    url = "https://niceweather.co.kr/product/" + uni
    page = requests.get(url)

    soup = bs(page.text, "html.parser")
    ele_rep_img = soup.select('div.swiper-slide > div:nth-child(1)')
    cnt = 0
    for i in ele_rep_img:
        rep_img = 'https://contents.sixshop.com'+i['imgsrc']
        if cnt == 0:
            if os.path.isdir(f"big/{uni_dir}") == False:
                os.mkdir(f"big/{uni_dir}")
            urllib.request.urlretrieve(rep_img,'big/{}/rep_{}_{}.jpg'.format(uni_dir,uni,cnt))
        else:
            if os.path.isdir(f"excel/old_rep/{uni_dir}") == False:
                os.mkdir(f"excel/old_rep/{uni_dir}")
            urllib.request.urlretrieve(rep_img,'excel/old_rep/{}/rep_{}_{}.jpg'.format(uni_dir,uni,cnt))
        cnt += 1
    count += 1
    print(f'{count}/ {len(uni_list)}')
    li.append('web/upload/old/rep/{}/rep_{}_{}.jpg'.format(uni_dir,uni,cnt))
print("fin.")
print(li)
```
